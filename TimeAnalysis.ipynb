# Analysis of time requirements for RSA key operations.


*   Key generation
*   Encryption
*   Decryption





```
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```
#time(key generation, decryption, encryption) taken for five differnt key sizes, obtained from timeAnalysisRSA.java
key_size = np.array([1024, 2*1024, 3*1024, 4*1024, 5*1024]) #in bits
gen_time = np.array([38200, 162500, 557000, 2393300, 3086500]) #times are taken avg. over 10 runs
dec_time = np.array([1900, 5200, 9700, 15100, 24300])
enc_time = np.array([200, 500, 1000, 1100, 1100])
```

```
plt.plot(key_size, gen_time, 'r-', marker='o',label="Key generation")
plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA key generation as a function of key size ");
```

```
plt.plot(key_size, dec_time, 'g-', marker='o')
plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA decrytion as a function of key size ");
```

```
plt.plot(key_size, enc_time,'b-', marker='o')
plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA encryption as a function of key size ");
```

```
#overall
#plt.rcParams["figure.figsize"] = (10,10)

plt.plot(key_size, gen_time, 'r-', label="Key generation") #linestyle='dashed')
plt.plot(key_size, dec_time,'g-', label="Decryption")
plt.plot(key_size, enc_time, 'b-', label="Encryption")

#create legend
plt.legend(loc="upper left")

plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA key operations as a function of key size ");
```

```
#overall using logarithm scale
plt.yscale("log")
plt.tick_params(axis='both', which='major', labelsize=14)
plt.rcParams["figure.figsize"] = (8,8)

plt.plot(key_size, gen_time, 'r-', label="Key generation", marker='o')
plt.plot(key_size, dec_time,'g-', label="Decryption", marker='o')
plt.plot(key_size, enc_time, 'b-', label="Encryption", marker='o')

#create legend
plt.legend(loc="upper left")

#labeling
plt.xlabel('Key size', fontsize=14)
plt.ylabel('Time (microseconds)', fontsize=14)
plt.title("Time for RSA key operations as a function of key size ", fontsize=14);

plt.savefig('my_figure.png')
```
