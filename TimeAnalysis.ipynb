```
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```
#time(key generation, decryption, encryption) taken for five differnt key values
key = np.array([1024, 2*1024, 3*1024, 4*1024, 5*1024])
gen_time = np.array([206000, 358000, 924000, 1777000, 1744000])
dec_time = np.array([6000, 25000, 13000, 28000, 57000])
enc_time = np.array([1000, 1000, 1000, 1000, 1000])
```

```
plt.plot(key, gen_time, 'r-', marker='o',label="Key generation")
plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA key generation as a function of key size ");
```

```
plt.plot(key, dec_time, 'g-', marker='o')
plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA decrytion as a function of key size ");
```

```
plt.plot(key, enc_time,'b-', marker='o')
plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA encryption as a function of key size ");
```

```
#overall
plt.plot(key, gen_time, 'r-', label="Key generation") #linestyle='dashed')
plt.plot(key, dec_time,'g-', label="Decryption")
plt.plot(key, enc_time, 'b-', label="Encryption")

#create legend
plt.legend(loc="upper left")

plt.xlabel('Key size')
plt.ylabel('Time (microseconds)')
plt.title("Time for RSA key operations as a function of key size ");
```

```
#overall using logarithm scale
plt.yscale("log")
plt.tick_params(axis='both', which='major', labelsize=14)
#plt.yticks(fontsize=14)# rotation=90)
plt.rcParams["figure.figsize"] = (10,10)

plt.plot(key, gen_time, 'r-', label="Generation", marker='o')# linewidth=3, markersize=12
plt.plot(key, dec_time,'g-', label="Decryption", marker='o')
plt.plot(key, enc_time, 'b-', label="Encryption", marker='o')

#create legend
plt.legend(loc="upper left")

#labeling
plt.xlabel('Key size', fontsize=20)
plt.ylabel('Time (microseconds)', fontsize=20)
plt.title("Time for RSA key operations as a function of key size ", fontsize=20);

plt.savefig('my_figure.png')
```
